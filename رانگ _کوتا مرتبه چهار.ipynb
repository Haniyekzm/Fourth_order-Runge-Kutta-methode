```python
 #فراخوانی کتابخانه ها
import matplotlib.pyplot as plt
import math
import numpy as np
import sympy as sp

#dy/dx= -g*t^2 + V0*t
#f(x,y)=dx/dy
#get the initial velocity 
v0=float(input("Enter the initial value of velocity :"))
#t0=float(input("Enter the initial value of y :"))
g=9.8
#تعریف تابع و دامنه و برد تابع و رسم ان
def f(t,y):
    return -g*t+v0
    
y=np.linspace(0,100,1000)

#plt.show()
#plt.xlabel('t')
#plt.ylabel('f(t,y)')
#plt.title('')
#plt.legend()
#plt.plot(f(t,y))

#When v=dy/dx=0 H is max so v=-g*t+v0=0 ==>
tfinal=v0/g
#Get the initial values of t ,y,h 
t1=float(input("Enter the initial value of y :"))
y1=float(input("Enter the initial value of t :"))
h=float(input("Enter the h :"))
tf=t1+tfinal

#print("-----=",f(t1,y1))
#i=1
#print(":))99)))=",i)
fig =plt.figure()
plt.xlabel('time')
plt.ylabel('height')
ax=plt.subplot()
i=0
while t1<=tf:
    s1=f(t1,y1)
    s2=f(t1+h/2,y1+s1*h/2)
    s3=f(t1+h/2,y1+s2*h/2)
    s4=f(t1+h,y1+s3*h)
    s=s1+2*s2+2*s3+s4
    y2=y1+s*h/6
    t1=t1+h
    y1=y2
    ax.scatter(t1,y1)
    i+=1
    if i>5000:
        break
print(" the distance the ball travels until its speed reaches zero for fist time is :",y1)
#print("&&&&&&=",t1)
plt.show()
```
